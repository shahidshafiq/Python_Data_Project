```python
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt  

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
import ast
df['job_skills'] = df['job_skills'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x)
```

```python
df["job_work_from_home"].value_counts().plot(kind = "pie", startangle = 90, autopct = "%1.1f%%")
plt.ylabel("")
plt.title("WORKING FROM HOME DATA")
```

```python
df["job_title_short"].value_counts().plot(kind = "pie", autopct = "%1.1f%%")
```

```python
df[["job_work_from_home", "job_no_degree_mention", "job_health_insurance"]]
```

```python
fig, ax = plt.subplots(1,3)
dict_column = {
    "job_work_from_home" : "work from home",
    "job_no_degree_mention" : "no degree required",
    "job_health_insurance": "health insurance",
    }
for i, (key,value) in enumerate(dict_column.items()):
    df[key].value_counts().plot(kind = "pie", ax = ax[i], startangle = 90, autopct = "%1.1f%%")
    #ax[i].pie(df[key].value_counts())
    ax[i].set_title(value)  
    ax[i].set_ylabel("")          
```
