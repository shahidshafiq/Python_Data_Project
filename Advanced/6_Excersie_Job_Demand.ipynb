```python
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt  

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
df_India = df[df["job_country"] == "India"].copy()
df_India["job_posted_month"] = df_India["job_posted_date"].dt.strftime("%B")
df_India


```

```python
dd = df_India.pivot_table(index= "job_posted_month", columns= "job_title_short", aggfunc= "size")
```

```python
dd
```

```python
dd.reset_index(inplace = True)
```

```python
dd
```

```python
dd["job_posted_month_num"] = pd.to_datetime(dd["job_posted_month"], format = "%B").dt.month
```

```python
dd
```

```python
dd.sort_values( "job_posted_month_num", inplace = True)
```

```python
dd.set_index("job_posted_month", inplace = True)
```

```python
dd.drop(columns= "job_posted_month_num", inplace = True)
```

```python
toop_3 = df_India["job_title_short"].value_counts().head(3).index.tolist()
```

```python
dd[toop_3].plot(kind= "bar")
plt.title("Jobs in India")
plt.ylabel("Number of Jobs")
plt.xlabel("2023")
plt.xticks(rotation = 360)
plt.legend()
plt.show()
```
