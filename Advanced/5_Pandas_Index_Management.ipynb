```python
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt  

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
df.sample(4)
```

```python
df.index.name = "job_index"
df.head(3)
```

```python
df_usa = df[df["job_country"] == "United States"]
df_usa.reset_index(inplace=True)
df_usa
```

```python
df_usa.set_index("job_index", inplace = True)
df_usa
```

```python
median_pivot = df.pivot_table(values='salary_year_avg', index='job_title_short', aggfunc=('median', "mean"))
median_pivot.sort_values(by = "job_title_short", inplace = True)
median_pivot
```
