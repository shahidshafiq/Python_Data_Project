```python
import pandas as pd
import matplotlib.pyplot as plt
from datasets import load_dataset
dataset = load_dataset("lukebarousse/data_jobs")
df = dataset["train"].to_pandas()
df["job_posted_date"] = pd.to_datetime(df["job_posted_date"])
```

```python
import ast
df["job_skills"] = df["job_skills"].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x)
```

```python
df_da = df[(df["job_title_short"]== "Data Scientist") & (df["job_country"] == "United States")].copy()
```

```python
df_da_clean = df_da.dropna(subset = ["salary_year_avg"])
```

```python
df_da_clean["salary_year_avg"].plot(kind = "hist", bins = 50 , edgecolor = "white")
plt.xlim(30000, 300000)
plt.ylim(0,1000)
plt.xlabel("Salary")
plt.ylabel("No of Jobs Advertized")
plt.title("Data Scientist Job Posted Across The United States")
```
