From Dataset We will be Exploring The Data Scientist Jobs Posted Across India


```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from datasets import load_dataset
dataset = load_dataset("lukebarousse/data_jobs")
df = dataset["train"].to_pandas()
df["job_posted_date"] = pd.to_datetime(df["job_posted_date"])
import ast
df["job_skills"] = df["job_skills"].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x)
# Only get data scientist jobs in the India
df_DS_IND = df[(df['job_title_short'] == 'Data Scientist') & (df['job_country'] == 'India')].copy()
```

```python
df_DS_INDIA = df_DS_IND["job_location"].value_counts().to_frame()
```

```python
df_DS_INDIA = df_DS_INDIA.head(10)
```

```python
sns.set_theme(style="ticks")
sns.barplot(data = df_DS_INDIA, x = "count", y = "job_location", hue= "count", palette= "dark:b_r", legend=False)
plt.title("Data Scientist Jobs Posting Across India")
plt.xlabel("Number of Jobs")
plt.ylabel("")
plt.show()
```

```python
fig, ax = plt.subplots(1,3)
fig.set_size_inches(12,6)
dict_column = {
    "job_work_from_home" : "Work from Home",
    "job_no_degree_mention" : "No Degree Required",
    "job_health_insurance": "Health Insurance",
    }
for i, (key,value) in enumerate(dict_column.items()):
    df_DS_IND[key].value_counts().plot(kind = "pie", ax = ax[i], startangle = 90, autopct = "%1.1f%%")
    #ax[i].pie(df[key].value_counts())
    ax[i].set_title(value)  
    ax[i].set_ylabel("")         
```

```python
df_DS_INDIA = df_DS_IND["company_name"].value_counts().to_frame()
df_DS_INDIA = df_DS_INDIA.head(10)
sns.set_theme(style="ticks")
sns.barplot(data = df_DS_INDIA, x = "count", y = "company_name", hue= "count", palette= "dark:b_r", legend=False)
plt.title("Companies Posted Data Scientist Across India")
plt.xlabel("Number of Jobs")
plt.ylabel("")
plt.show()

```
